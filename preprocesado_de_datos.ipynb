```python
import pandas as pd

# Asumiendo que 'df' es tu DataFrame original

# Primero, cambiamos el formato del DataFrame a un formato largo
df_long = df.melt(id_vars=['ID', 'Consumo (0) / Producción (1)', 'Dia', 'Mes ', 'Año'], 
                  var_name='Hora', value_name='Lectura')

# Convertimos la columna 'Hora' a un entero
df_long['Hora'] = df_long['Hora'].astype(int)

# Creamos un diccionario para mapear los intervalos a horas del día
interval_to_hour = {i: f'{(i-1)//4}:{15*((i-1)%4):02}' for i in range(1, 97)}

# Aplicamos el mapeo a la columna 'Hora'
df_long['Hora'] = df_long['Hora'].map(interval_to_hour)

df_long['Hora'] = pd.to_datetime(df_long['Hora'], format='%H:%M').dt.time

# Ordenamos el DataFrame por ID y fecha
df_long.sort_values(by=['ID', 'Año', 'Mes ', 'Dia', 'Hora'], inplace=True)

# Restablecemos el índice
df_long.reset_index(drop=True, inplace=True)
```

```python
produccion = df_long[df_long['Consumo (0) / Producción (1)'] == 1]
consumo = df_long[df_long['Consumo (0) / Producción (1)'] == 0]
```

```python

# Tras revisar que el DataFrame esta ordenado de la manera deseada
df_long = df_long.reset_index()
df_long = df_long.rename(columns={'index': 'EntryID'})
df_long['EntryID'] = df_long['EntryID'] + 1
df_long
```

```python
df = df_long.rename(columns={
    'ID': 'ID',
    'Consumo (0) / Producción (1)': 'ConsumptionOrProduction',
    'Dia': 'Day',
    'Mes ': 'Month',
    'Año': 'Year',
    'Hora': 'Time',
    'Lectura': 'Reading'
})
```
